```python



df[' Price '] = pd.to_numeric(df[' Price '].replace('[\$,]', '', regex=True), errors='coerce')

# Get the top 10 coins based on price
top_10_coins = df.nlargest(10, ' Price ')

top_10_coins= top_10_coins.set_index('Rank')


print(top_10_coins[['Coin Name', ' Price ']])
```

```python


df[' Price '] = pd.to_numeric(df[' Price '].replace('[\$,]', '', regex=True), errors='coerce')

df_cleaned_price = df.dropna(subset=[' Price ']).drop_duplicates(subset=[' Price '])

# Get the bottom 10 unique coins based on price
bottom_10_coins_price = df_cleaned_price.nsmallest(10, ' Price ')


bottom_10_coins_price = bottom_10_coins_price.set_index('Rank')

print(bottom_10_coins_price[['Coin Name', ' Price ']])
```
